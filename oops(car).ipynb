```python
class car:
    
    purpose='to travel and visit the place which we wish to'
    
    def __init__(self,brand,color,model):
        self.brand=brand
        self.color=color
        self.model=model
        self.currentfuel=3
        self.minfuel=1
        self.maxfuel=37
        self.gear=0
        self.maxgear=5
        self.speed=0
        self.key=False
        
    def start(self):
        if self.key==False:
            self.key=True
            print(self.brand ,self.model,'is started and drive safely')
        else:
            print(self.brand ,self.model,'is already started. Good to go')
    
    def stop(self):
        if self.key==False:
            print(self.brand ,self.model,'is already off.')
        else:
            self.key=False
            print(self.brand ,self.model,'is off and see you soon')
    
    def gearinc(self):
        if self.key==True:
            self.gear+=1
            if self.gear>self.maxgear:
                print('max availability is done')
            else:
                print('current gear after increased',self.gear)
        else:
            print('please start the', self.brand ,self.model,'car')
            
    def geardec(self):
        if self.key==True:
            self.gear-=1
            if self.gear<0:
                print('warning. . it is not possible')
            elif self.gear==0:
                print('you are at neutral')
            else:
                print('current gear after decrement',self.gear)
        else:
            print('please start the', self.brand ,self.model,'car') 
            
    def fuel(self):
        if self.currentfuel<=self.minfuel:
            print('please fill the fuel tank')
            print('current fuel is',self.currentfuel)
        else:
             print('current fuel is',self.currentfuel)
        
        
    def fueladd(self):
        if self.key==False:
            if self.currentfuel>self.maxfuel:
                print('max availability is done')
            else:
                self.currentfuel+=2
                print('currrent fuel is',self.currentfuel)
        else:
            print('please of the',self.brand,self.model,'car to fill the tank')
            
            
    def incspeed(self):
        if self.key==True:
            self.speed+=10
            print('current speed is',self.speed)
            if self.speed>80:
                print('you have reached max limit and drive carefully')       
        else:
             print('please start the', self.brand ,self.model,'car') 
                
    def decspeed(self):
        if self.key==True:
            self.speed-=10
            if self.speed<0:
                print('car is stopped. No more decrease is possible')
            else:
                print('currrent speed is',self.speed)        
        else:
             print('please start the', self.brand ,self.model,'car') 
    
    
    def nightmode(self):
        if self.key==True:
            print('switch on the headlight')
        else:
            print('switch on the headlight')
            print('but', self.brand ,self.model,'car is off') 
    
    def getdetails(self):
        if self.key==True:
            print('current deatils : ')
            print('current fuel is',self.currentfuel)
            print('current gear is',self.gear)
            print('current speed is',self.speed)
            
        else:
            print('please start the', self.brand ,self.model,'car')
            
```
